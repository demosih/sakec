```python
 import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
 X = np.array([[[i], [i+1], [i+2]] for i in range(10)])  # shape (10,3,1)
y = np.array([i+3 for i in range(10)])                  # next value prediction

```

```python
model = Sequential([
    LSTM(50, input_shape=(3,1)),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X, y, epochs=100, verbose=0)
```

```python
test = np.array([[[10], [11], [12]]])
print("Predicted next value:", model.predict(test, verbose=0))
```
